```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
# Read the City and Ride Data
city_data_to_load = pd.read_csv("data/city_data.csv")
ride_data_to_load = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
combined_data = pd.merge(city_data_to_load, ride_data_to_load, on="city")

# Display the data table for preview
combined_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_groups = combined_data.groupby('city')

city_groups_data = {'Number of Rides': city_groups.size(),
                    'Fare': city_groups['fare'].mean(),
                    'Type': city_groups['type'].first(),
                    'Driver Count': city_groups['driver_count'].max()*10}

## Plot the charts and apply some styling
# Incorporate a text label regarding circle size
plt.scatter('Number of Rides',
            'Fare',
            s='Driver Count',
            c=('gold', 'lightskyblue', 'lightcoral'),
            data=city_groups_data,
            alpha=0.5)

# Create a legend
lgnd = plt.legend()

plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare ($)")


# Incorporate the other graph properties
plt.grid(True)


# Build the scatter plots for each city types..


# Save Figure


# group = pd.DataFrame(data=city_groups_data)


# plt.scatter('Number of Rides',
#             'Fare',
#            s='Driver Count',
#             c=['yellow', 'blue', 'red'],
#             cmap='Type',
#             data=city_groups_data,
#             alpha=0.5)

#Urban, = plt.plot(time, speed_chain, color="red", label="Danger Drop")
#Suburban, = plt.plot(time, speed_launch, color="blue", label="RailGun")
#Rural, = plt.plot(time, speed_launch, color="blue", label="RailGun")

#plt.legend(loc = 'upper right')
#plt.legend(title="City Types", loc='upper right')
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_groups = combined_data.groupby('type')
type_groups_data = {'Fare': type_groups['fare'].sum()}
type_groups_data

#Labels for the sections of our pie chart
labels = ['Urban', 'Suburban', 'Rural']

#The values of each section of the pie chart
sizes = [39854.38, 19356.33, 4327.93]

#The colors of each section of the pie chart
colors = ['lightcoral', 'lightskyblue', 'gold']

#Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0.2, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")

# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
type_groups = combined_data.groupby('type')
type_groups_data = {'Number of Rides': type_groups.count()}
type_groups_data
# Build Pie Chart
#Labels for the sections of our pie chart
labels = ['Urban', 'Suburban', 'Rural']

#The values of each section of the pie chart
sizes = [1625, 625, 125]

#The colors of each section of the pie chart
colors = ['lightcoral', 'lightskyblue', 'gold']

#Tells matplotlib to seperate the "Python" section from the others
explode = (0.1, 0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")

# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

ride_percents = 100 * combined_data.groupby(["type"]).sum()["driver_count"] / combined_data["driver_count"].sum()

# Build Pie Chart
plt.pie(ride_percents, 
        labels=["Rural", "Suburban", "Urban"], 
        colors=["gold", "lightskyblue", "lightcoral"], 
        explode=[0, 0, 0.1], 
        autopct='%1.1f%%', 
        shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")

# Save Figure
#plt.savefig("analysis/Fig3.png")

# Show Figure
plt.show()



#type_groups = combined_data.groupby('type')
#type_groups_data = {'Driver Count': type_groups['driver_count'].sum()}
#type_groups_data
# Build Pie Charts

#Labels for the sections of our pie chart
#labels = ['Urban', 'Suburban', 'Rural']

#The values of each section of the pie chart
#sizes = [59602, 8570, 537]

#The colors of each section of the pie chart
#colors = ['red', 'blue', 'yellow']

#Tells matplotlib to seperate the "Python" section from the others
#explode = (0.1, 0.2, 0)

#plt.pie(sizes, explode=explode, labels=labels, colors=colors,
#        autopct="%1.1f%%", shadow=True, startangle=140)
#plt.axis("equal")
# Save Figure



```
